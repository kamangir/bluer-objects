```python
import numpy as np
import random

from bluer_options import string
from bluer_sandbox.notebooks import imshow
from bluer_objects.logger.stitch import stitch_images

from bluer_sandbox.host import signature
from bluer_sandbox.logger import logger

logger.info(f"{'.'.join(signature())},\nbuilt on {string.pretty_date()}")
```

---

```python
count = 10
rng = np.random.default_rng()
list_of_images = [
       rng.integers(
           0, 
           256, 
           (
               random.randint(1, 512), 
               random.randint(1, 512), 
               random.choice([1, 3])
            ), 
            dtype=np.uint8
        ) for _ in range(count)
]
```

```python
image = stitch_images(
    list_of_images, 
)

logger.info(f"image: {string.pretty_shape_of_matrix(image)}")
```

```python
imshow(image)
```
