## `search`

```python
value = string.timestamp()
logger.info(f"value: {value}")
```

```python
for index in range(3):
    object_name = objects.unique_object(f"test-mlflow-search-{index}", log=False)

    set_tags(
        object_name=object_name,
        tags={
            "this": value,
            "index": str(index),
        },
        verbose=False,
    )
```

```python
success, list_of_objects = search(
    filter_string=f"this={value},index=2",
    verbose=False,
    )
assert success
log_list(logger,"list of",list_of_objects)
```
